**Task 1**: Let's look at the relationship between fuel mileage ratings for city vs. highway driving, as stored in the 'city' and 'highway' variables (in miles per gallon, or mpg). Use a _scatter plot_ to depict the data. What is the general relationship between these variables? Are there any points that appear unusual against these trends?

```python
# make a larger figure
plt.figure(figsize = [16, 5])

# create scatter plot with matplotlib
plt.subplot(1, 3, 1)
plt.scatter(data = fuel_econ, x = 'city', y = 'highway', alpha = 0.1)
plt.title('Matplotlib - scatter')
plt.xlabel('city (mpg)')
plt.ylabel ('highway (mpg)')

# create an regplot with seaborn
plt.subplot(1, 3, 2)
sb_color = sb.color_palette()[3]
sb.regplot(data = fuel_econ, x = 'city', y = 'highway',fit_reg = False, color = sb_color, x_jitter = 0.2,
          y_jitter = 0.2, scatter_kws = {'alpha': 1/6})
plt.title('Seaborn - regplot')
plt.xlabel('city (mpg)')
plt.ylabel ('highway (mpg)');

# create a heatmap with hist2d
plt.subplot(1, 3, 3)
bins_x = np.arange(fuel_econ['city'].min() -1, fuel_econ['city'].max()+1, 1)
bins_y = np.arange(fuel_econ['highway'].min() -1, fuel_econ['highway'].max()+1, 1)
plt.hist2d(data = fuel_econ, x = 'city', y = 'highway', cmap = 'BuPu_r',cmin =0.5, bins =[bins_x, bins_y])
plt.title('Matplotlib - hist2d')
plt.colorbar(orientation='horizontal', fraction= 0.05)
plt.xlabel('city (mpg)')
plt.ylabel ('highway (mpg)');
```

**Task 2**: Let's look at the relationship between two other numeric variables. How does the engine size relate to a car's CO2 footprint? The 'displ' variable has the former (in liters), while the 'co2' variable has the latter (in grams per mile). Use a heat map to depict the data. How strong is this trend?

```python
# create a larger figure 
plt.figure(figsize = [16, 5])

# subplot 1 with scatter
plt.subplot(1, 3, 1)
plt.scatter(data = fuel_econ, x = 'displ', y = 'co2', alpha = 0.1)
plt.title('Matplotlib - scatter')
plt.xlabel('engine size (L)')
plt.ylabel('CO2 (g/mile)')

# subplot 2 with regplot
plt.subplot(1, 3, 2)
sb.regplot(data = fuel_econ, x ='displ', y = 'co2', x_jitter=0.2, y_jitter=0.2, fit_reg =False,
          scatter_kws={'alpha':0.1})
plt.title('Seaborn - regplot')
plt.xlabel('engine size (L)')
plt.ylabel('CO2 (g/mile)')

# subplot 3 with hist2d
plt.subplot(1, 3, 3)
bins_x = np.arange(0.6, 8, 0.5)
bins_y = np.arange(29, 693, 50)
plt.hist2d(data=fuel_econ, x='displ', y='co2', bins = [bins_x, bins_y], cmap='BuPu',
           cmin=0.5) # defin empty cell with cmin
plt.colorbar(orientation='horizontal', fraction= 0.05)
plt.title('Matplotlib - hist2d')
plt.xlabel('engine size (L)')
plt.ylabel('CO2 (g/mile)');
```
