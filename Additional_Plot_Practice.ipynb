**Task 1**: Practice creating a plot matrix, by depicting the relationship between five numeric variables in the fuel efficiency dataset: 'displ', 'co2', 'city', 'highway', and 'comb'. Do you see any interesting relationships that weren't highlighted previously?


#### We use seaborn pariGrid to create a plot matrix between those five variables.

```python
# make a copy of dataframe
fuel_econ1 = fuel_econ.copy()
```

```python
# create a plot of matrix with pairGrid function
g = sb.PairGrid(data = fuel_econ1, vars=['displ', 'co2', 'city', 'highway', 'comb'], diag_sharey=True,  despine=True, dropna=True)
base_color = sb.color_palette()[4]
g.map_diag(sb.kdeplot, color =base_color)
g.map_offdiag(sb.regplot, fit_reg = False, x_jitter = 0.2, y_jitter =0.2, color = base_color, scatter_kws = {"alpha": 0.2})
g.set(xscale ='log', yscale='log')

# add plot title
g.fig.subplots_adjust(top=0.95)
g.fig.suptitle('PairGrid plot matrix to examine five variables',fontsize=16);
```

**Task 2**: The output of the preceding task pointed out a potentially interesting relationship between co2 emissions and overall fuel efficiency. Engineer a new variable that depicts CO2 emissions as a function of gallons of gas (g / gal). (The 'co2' variable is in units of g / mi, and the 'comb' variable is in units of mi / gal.) Then, plot this new emissions variable against engine size ('displ') and fuel type ('fuelType'). For this task, compare not just Premium Gasoline and Regular Gasoline, but also Diesel fuel.

#### We need calculate CO2 emmisions

- 'co2' variable * 'comb' variable

- add new columns named as 'gallon_gas'

- create PairGrid to plot 'displ', 'fuelType' and 'gallon_gas'


```python
fuel_econ1['gallon_gas'] = fuel_econ1['co2'] * fuel_econ1['comb'].astype(int)
fuel_econ1.columns
```

```python
fuel_econ1.gallon_gas.head()
```

```python
# filter the fuel_type 'fuelType'containing a list of 'premium', 'regular' and 'diesel' stings
fuel_type = ["Premium Gasoline", "Regular Gasoline", "Diesel"]

# refer to variables in the environment by prefixing them with an '@' character like  @a + b
fuel_econ2 = fuel_econ1.query('fuelType == @fuel_type')
```

```python
# Facet plot
g = sb.FacetGrid(data = fuel_econ2, col = 'fuelType', hue = 'cylinders', sharey = True, height = 5, despine = True, dropna = True)
g.map(sb.scatterplot, 'gallon_gas', 'displ', edgecolor ='w', alpha = 'auto',
      x_jitter = 0.2, y_jitter = 0.2, legend = 'brief');
g.add_legend();

# add plot title
g.fig.subplots_adjust(top=0.8)
g.fig.suptitle('FacetGrid to plot 4 variables',fontsize=16);
 
```
