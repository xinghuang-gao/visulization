**Task 1**: Plot the city ('city') vs. highway ('highway') fuel efficiencies (both in mpg) for each vehicle class ('VClass'). Don't forget that vehicle class is an ordinal variable with levels {Minicompact Cars, Subcompact Cars, Compact Cars, Midsize Cars, Large Cars}.

```python
fuel_econ1 = fuel_econ.copy()
vclass_order = ["Minicompact Cars", "Subcompact Cars", "Compact Cars", "Midsize Cars", "Large Cars"]
g = sb.FacetGrid(data = fuel_econ1, col ='VClass', col_wrap = 3, col_order = vclass_order, sharey = True, height = 5, dropna = True )
g.map(sb.scatterplot, 'city', 'highway', color = 'purple', x_jitter =0.2, y_jitter = 0.2,
      alpha = 'auto', edgecolor = 'w');
# g.set(xscale ='log', yscale='log');


# add plot title
g.fig.subplots_adjust(top=0.85)
g.fig.suptitle('Scatter plot with 3 variables',fontsize=16);
g.set_titles("{col_name}", size = 16)
g.set_xlabels('City (mpg)', fontsize = 14)
g.set_ylabels('Highway (mpg)', fontsize = 14)
g.set_xticklabels(fontsize = 15);
```

**Task 2**: Plot the relationship between engine size ('displ', in meters), vehicle class, and fuel type ('fuelType'). For the lattermost feature, focus only on Premium Gasoline and Regular Gasoline cars. What kind of relationships can you spot in this plot?

```python
fuel_type = ["Premium Gasoline", "Regular Gasoline"]
fuel_econ2 = fuel_econ1.query('fuelType == @fuel_type')
vclass_order = ["Minicompact Cars", "Subcompact Cars", "Compact Cars", "Midsize Cars", "Large Cars"]

sb.boxplot(data = fuel_econ2, x = 'VClass', y= 'displ', hue = 'fuelType', palette='Set2', order = vclass_order)
plt.xticks(rotation =30)
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5));
```

```python
vclass_order = ["Minicompact Cars", "Subcompact Cars", "Compact Cars", "Midsize Cars", "Large Cars"]

g = sb.FacetGrid(data = fuel_econ2, col = 'VClass', col_order = vclass_order,
                 col_wrap =3, sharey = True, height = 3)
g.map_dataframe(sb.boxplot, 'fuelType', 'displ',
                palette='Set2', width = 0.4) # not work with g.map() function
g.set_titles("{col_name}", size = 16)
g.set_axis_labels('Fuel Type', 'Displ');
```
